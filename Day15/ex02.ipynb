```python
# step1. selenium 패키지와 time 모듈 import
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

# step2. 검색할 키워드 입력
query = input('검색할 키워드를 입력하세요: ')

# step3. 크롬드라이버로 원하는 url로 접속
url = 'https://www.naver.com/'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)

# step4. 검색창에 키워드 입력 후 엔터
search_box = driver.find_element(By.ID, "query")
search_box.send_keys(query)
search_box.send_keys(Keys.RETURN)
time.sleep(2)

# step5. 첫번째 탭 클릭 (CSS_SELECTOR 사용)
driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > div.lnb_nav_area._nav_area_root > div > div.api_flicking_wrap._conveyer_root > div:nth-child(1)').click()
time.sleep(2)
```

```python
#step6. 제목 텍스트 추출

titles = driver.find_elements(By.CLASS_NAME, "title_link")

for i in titles:
    title = i.text
    print(title)
```

```python
#step7.제목 하이퍼링크 추출

for i in titles:
    href = i.get_attribute('href')
    print(href)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
from urllib.request import urlretrieve

# step1. selenium 패키지 import
driver = webdriver.Chrome()

# step2. 네이버 블로그 검색
url = 'https://search.naver.com/search.naver?query=블로그'
driver.get(url)

# step3. 블로그 탭 클릭
driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > div.lnb_nav_area._nav_area_root > div > div.api_flicking_wrap._conveyer_root > div:nth-child(8)').click()
time.sleep(2)

# step4. 스크롤 내리기 (썸네일 이미지 로딩)
for _ in range(5):  # 스크롤을 여러 번 내리며 이미지 로딩
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)

# step5. 썸네일 이미지 추출
wait = WebDriverWait(driver, 20)  # 대기 시간을 늘림

# `thumb` 클래스를 사용하여 썸네일 이미지 영역을 기다림
news_content_div = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'news_contents')))  # 썸네일 영역을 찾음

news_thumbnail = []

for i in news_content_div:
    try:
        # 각 'news_contents' 안에서 img 태그 찾기
        thumbnails = i.find_elements(By.TAG_NAME, 'img')  # 'img' 태그들 찾기
        news_thumbnail.extend(thumbnails)  # 발견된 이미지들을 목록에 추가
    except Exception as e:
        print(f"썸네일 이미지 추출 오류: {e}")

# 이미지 링크 리스트 추출
link_thumbnail = []
for img in news_thumbnail:
    src = img.get_attribute('data-lazysrc') or img.get_attribute('src')  # 'data-lazysrc'가 없으면 'src' 사용
    if src:
        link_thumbnail.append(src)

# 추출된 이미지 링크 확인
print(f"추출된 썸네일 이미지 링크: {link_thumbnail}")

# 이미지 저장 폴더 생성
path_folder = 'C:/korea3/'

if not os.path.isdir(path_folder):
    os.mkdir(path_folder)

# step6. 이미지 다운로드
i = 0
for link in link_thumbnail:
    if link:  # URL이 존재하는 경우만 다운로드
        i += 1
        try:
            urlretrieve(link, os.path.join(path_folder, f'{i}.jpg'))
            print(f"{i}.jpg 저장 완료")
        except Exception as e:
            print(f"{i}.jpg 저장 실패: {e}")
    else:
        print(f"유효하지 않은 URL: {link}")
    time.sleep(0.3)

driver.quit()
```
